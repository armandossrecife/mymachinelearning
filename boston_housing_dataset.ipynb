# Boston dataset

The Boston Housing Prices dataset, loaded using `sklearn.datasets.load_boston()`, is a well-known dataset in machine learning for regression tasks. It provides information about the housing market in Boston, Massachusetts, and the goal is to predict housing prices based on various features that can influence them.

Here's a breakdown of the dataset:

**Data Structure:**

* When you call `load_boston()`, it returns a dictionary-like object containing the following key components:
    * `data`: This is a NumPy array with a shape of (number of samples, number of features). It holds the actual values for each feature across all data points (houses) in the dataset.
    * `target`: This is a NumPy array with a shape of (number of samples,). It contains the target variable, which is the median value of owner-occupied homes in thousands of dollars.
    * `feature_names`: This is a list of strings, where each string represents the name or description of a feature in the `data` array.
    * `DESCR`: This is a string containing a detailed description of the dataset, including information about the features and the source of the data.

**Features and Target Variable:**

* The dataset includes 13 features (predictors) that might influence housing prices and a single target variable:
    * **Features (predictors):**
        * CRIM: per capita crime rate by town
        * ZN: proportion of residential land zoned for large-lot construction
        * INDUS: proportion of non-retail business acres per town
        * CHAS: Charles River dummy variable (binary, 1 if tract borders Charles River)
        * NOX: nitric oxide concentration (parts per million)
        * RM: average number of rooms per dwelling
        * AGE: median age of owner-occupied units (built in)
        * DIS: weighted distance to five Boston employment centers
        * RAD: index of accessibility to radial highways
        * TAX: full-value property tax rate per $10,000
        * PTRATIO: pupil-teacher ratio by town
        * B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
        * LSTAT: % lower status of the population
    * **Target Variable:**
        * MEDV: median value of owner-occupied homes in $1000s

It's important to understand the meaning and potential impact of each feature on housing prices when working with this dataset. For instance, features like the crime rate (CRIM), the average number of rooms (RM), and the distance to employment centers (DIS) might be expected to have a negative or positive correlation with housing prices.

**Use Cases:**

The Boston Housing Prices dataset is commonly used for:

* **Regression modeling:** The primary application is to train regression models to predict housing prices based on the provided features. This can be helpful in various scenarios, like real estate valuation or understanding the factors that drive housing market trends.
* **Machine learning algorithm evaluation:** Due to its well-understood nature and availability, the dataset serves as a benchmark for evaluating the performance of different regression algorithms. This allows researchers and practitioners to compare the effectiveness of various models.
* **Exploratory data analysis (EDA):**  The dataset can be used to explore relationships between features and the target variable through techniques like correlation analysis and visualization. This can provide insights into the factors that influence housing prices.

```python
from sklearn.datasets import load_boston
import pandas as pd

# Load the Boston housing prices dataset
boston = load_boston()

# Create a DataFrame using the feature names and data
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# Add the target variable (median value of owner-occupied homes) as a column
df['MEDV'] = boston.target
```

```python
# Print the first few rows of the DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt

# Choose a feature (e.g., RM - average number of rooms)
chosen_feature = 'RM'

# Create the scatter plot
plt.scatter(df[chosen_feature], df['MEDV'])  # MEDV is the median housing price

# Label the axes
plt.xlabel(chosen_feature)
plt.ylabel('MEDV (Median Housing Price)')

# Add a title
plt.title(f'Scatter Plot: {chosen_feature} vs. MEDV')

# Gridlines for better readability (optional)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Target variable (median housing price)
price_column = 'MEDV'

# Create the histogram
plt.hist(df[price_column], bins=20, edgecolor='black')  # Adjust bins as needed

# Label the axes
plt.xlabel('MEDV (Median Housing Price)')
plt.ylabel('Number of Houses')

# Add a title
plt.title('Distribution of Median Housing Prices')

# Show the plot
plt.show()
```

```python
# Example price ranges (bins)
price_ranges = [0, 30, 50, float('inf')]  # Adjust bin edges as needed

price_groups = pd.cut(df['MEDV'], bins=price_ranges, labels=range(len(price_ranges) - 1))
df['price_groups'] = price_groups
```

```python
import seaborn as sns

# Choose a feature (e.g., RM - average number of rooms)
chosen_feature = 'RM'

# Create the boxplot
sns.boxplot(
    x = "price_groups",
    y = chosen_feature,
    showmeans=True,  # Show mean as a diamond
    data=df
)

# Rotate x labels for better readability (optional)
plt.xticks(rotation=45)

# Label the axes and add a title
plt.xlabel('Price Range')
plt.ylabel(chosen_feature)
plt.title(f'Boxplots: {chosen_feature} Distribution by Price Range')

# Show the plot
plt.show()
```

```python
import seaborn as sns

correlation_matrix = df.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')  # Adjust colors as needed

# Label the axes
plt.xlabel('Features')
plt.ylabel('Features')

# Add a title
plt.title('Correlation Matrix (Heatmap)')

# Show the plot
plt.show()
```
